```python
from pydataset import data
import pandas as pd
import numpy as np

# data('mpg', show_doc=True) # view the documentation for the dataset
mpg = data('mpg') # load the dataset and store it in a variable

# All the datasets loaded from the pydataset library will be pandas dataframes.

# 1. Copy the code from the lesson to create a dataframe full of student grades.


np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
# a. Create a column named passing_english that indicates whether each student has a passing grade in english.

df.assign(passing_english=df.english >= 75)
df = df.assign(passing_english=df.english >= 75)
```

```python
# b.Sort the english grades by the passing_english column. How are duplicates handled?

df.sort_values(by='english')
```

```python
# c. Sort the english grades first by passing_english and then by student name.
# All the students that are failing english should be first, 
#and within the students that are failing english they should be ordered alphabetically. 
#The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)

df.sort_values(by = ['passing_english','name'])
```

```python
# d. Sort the english grades first by passing_english, 
# and then by the actual english grade, similar to how we did in the last step.

df.sort_values(by = ['passing_english', 'english'])
```

```python
# e. Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.
df = df.assign(passing_english=df.english >= 75)

columns = ['math', 'english', 'reading']
averages = df[columns].mean(axis = 1).round(1)
df['overal_grade'] = averages
df
```

```python
# . Rename the cty column to city.

mpg.rename(columns = {'cty' : 'city'}, inplace = True)
mpg
```

```python
# . Rename the hwy column to highway.

mpg.rename(columns = {'hwy' : 'highway'}, inplace = True)
mpg
```

```python
# . Do any cars have better city mileage than highway mileage?

mpg[mpg['city'] > mpg ['highway']]
```

```python
# . Create a column named mileage_difference this column should contain the 
# difference between highway and city mileage for each car.

mpg['mileage_difference'] = mpg['highway'] - mpg['city']
mpg

```

```python
# . Which car (or cars) has the highest mileage difference?

mpg.nlargest(n = 1, columns = 'mileage_difference')
```

```python
# . Which compact class car has the lowest highway mileage? The best?

mpg.nsmallest(n = 1, columns = 'mileage_difference')
```

```python
# . Create a column named average_mileage that is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg.city + mpg.highway)/2
mpg

```

```python
# . Which dodge car has the best average mileage? 

mpg[mpg.manufacturer == 'dodge'].nlargest(n = 1, columns = 'average_mileage', keep = 'all')
```

```python
# The worst?

mpg[mpg.manufacturer == 'dodge'].nsmallest(n = 1, columns = 'average_mileage', keep = 'all')
```

```python
# . What is the the weight of the fastest animal?

mammals[mammals.speed == mammals.speed.max()].weight
```

```python
# . What is the overal percentage of specials?

str(mammals['specials'].mean().round(2) * 100) + '%'
```

```python
# . How many animals are hoppers that are above the median speed? 
hoppers = mammals[mammals.hoppers]
hoppers_above_median_speed = hoppers[hoppers.speed > mammals.speed.median()]
hoppers_above_median_speed.shape[0]
```

```python
# . What percentage is this?
str(round((hoppers_above_median_speed.shape[0] / mammals.shape[0]) * 100, 1)) + '%'
```
