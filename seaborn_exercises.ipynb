```python
# 4.Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items 
#and the revenue produced by each.

sql = '''
SELECT
    item_name,
    SUM(quantity) AS count,
    ROUND(SUM(CAST(SUBSTR(item_price, 2) AS FLOAT)), 2) AS revenue
FROM orders
GROUP BY item_name
ORDER BY revenue DESC
LIMIT 4;
'''

chipotle_popular_items = pd.read_sql(sql, get_db_url('chipotle'))
chipotle_popular_items
```

```python
plt.figure(figsize = (12, 6))
sns.barplot(data = chipotle_popular_items, x = 'item_name', y = 'revenue')
plt.title('Chipotle Most Popular Items');
```
