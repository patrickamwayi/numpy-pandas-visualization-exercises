```python
# Exercises

import matplotlib.pyplot as plt

# Do your work for this exercise in a jupyter notebook named matplotlib_exercises.

# 1. Use matplotlib to plot the following equation:

#y=x2 −x +2
#You'll need to write the code that generates the x and y points.

#Add an anotation for the point 0, 0, the origin.

#y = x** −x +2

x = list(range(-20,30))
y = [n* n - n+ 2 for n in x]
plt.plot(x, y)

plt.title('y = x^2 - x + 2')


plt.show()
```

```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

# y=√x
# y=x^3
# y=2^x
# y=1/(x+1)
# You can use functions from the math module to help implement some of the equations above.
# some data to play with

#      y = √x 

x = list(range(0, 50))
y = [(i**.3) for i in x]
plt.plot(x,y)
plt.show()

```

```python
#    y=2^x

x = list(range(0,50))
y = [(i**i)for i in x]

plt.plot(x,y)
plt.show()

```

```python
# 3 Combine the figures you created in the last step into one large figure with 4 subplots.

plt.figure(figsize=(12,8))
x = list(range(0,50))
y = [(i**.3) for i in x]
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('square root of x')

x = list(range(0, 50))
y = [(i**3) for i in x]
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('x^3')
         
x = list(range(0,50))
y = [(i**i)for i in x]
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('n^x')
         
x = list(range(0, 50))
y = [(1/ (i + 1)) for i in x ]
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('1/(x-1)')

plt.suptitle('The four functions')
plt.show()


```

```python
#. 4 Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. 
# Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize=(12,8))

x = list(range(0,50))
y = [(i**.3) for i in x]
plt.plot(x,y, c='blue')

x = list(range(0, 50))
y = [(i**3) for i in x]
plt.plot(x,y, c='red')
         
x = list(range(0,50))
y = [(i**i)for i in x]
plt.plot(x,y, c='green')
         
x = list(range(0, 50))
y = [(1/ (i + 1)) for i in x ]
plt.plot(x,y, c='black')

plt.xlim(0,9)
plt.ylim(0,10)

plt.title('The four functions')
plt.legend(['square root of x', 'x^3', 'n^x', '1/(x-1)' ])
plt.show()
```
